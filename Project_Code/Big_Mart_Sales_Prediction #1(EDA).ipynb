```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

test=pd.read_csv("Test.csv")
train=pd.read_csv("Train.csv")

# print("Train Data set Dimension: ",train.shape)
# print("Test Data set Dimension: ",test.shape)

# print("Train Data Attribuetes: ",train.columns)
# print("Test Data Attribuetes: ",test.columns)

# print("\nTrain Data Header:\n ",train.head())
# print("\nTest Data Header: \n",test.head())


combine_data=test+train
# print("Combine Data :\n",combine_data)
sale=train['Item_Outlet_Sales']


```

# Univariate EDA

```python
plt.hist(sale,color='green',bins=100, rwidth=1.5)
plt.xlabel("Item_Outlet_Sales")
plt.ylabel("count")
```
